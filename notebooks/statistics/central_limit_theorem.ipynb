```python
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_of_sample = [1, 5, 15, 30, 45, 60, 75, 90, 100]
```

```python
def plot_func(mean_list, m, n):
    k = 0
    fig, ax = plt.subplots(m, n, figsize=(8, 8))
    for i in range(0, m):
        for j in range(0, n):
            ax[i, j].hist(mean_list[k], 10, density=True)
            ax[i, j].set_title(label=num_of_sample[k])
            k = k + 1
    plt.show()
```

## Discrete unifrom distribution

```python
x_val = np.random.randint(-60, 60, 1000)
```

```python
plt.hist(x_val);
```

```python
# Central limit Theorem
means = []
for j in num_of_sample:
    x = [np.mean(
        np.random.randint(
            -60, 60, j)) for _i in range(1000)]
    means.append(x)
```

```python
plot_func(means, 3, 3)
```

```python
# exponentail distributed
# assume on population
# assume rate = 0.125
rate = 0.125
mu = 1/rate
sd = np.sqrt(1/rate**2)
```

```python
data_expon = expon.rvs(size=1000, scale=mu)
ax = sns.histplot(data_expon,
                  kde=True,
                  color='skyblue')
ax.set(xlabel='Exponential Distribution', ylabel='Frequency');
```

```python
exp_mean = []
for i in num_of_sample:
    x = [np.mean(np.random.exponential(mu, i)) for j in range(100)]
    exp_mean.append(x)
```

```python
plot_func(exp_mean, 3, 3)
```
