```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_of_sample = [1,  60, 75, 100]
means = []
```

```python
for j in num_of_sample:
    x = [np.mean(
        np.random.randint(
            -60, 60, j)) for _i in range(1000)]
    means.append(x)
```

```python
k = 0
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
for i in range(0, 2):
    for j in range(0, 2):
        # Histogram for each x stored in means
        ax[i, j].hist(means[k], 10, density=True)
        ax[i, j].set_title(label=num_of_sample[k])
        k = k + 1
plt.show()
```
