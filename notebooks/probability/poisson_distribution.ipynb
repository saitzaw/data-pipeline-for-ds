```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import poisson
from IPython.display import Math, display
sns.set(color_codes=True)
sns.set(rc={'figure.figsize': (5, 5)})
```

## Formula Poisson(mu)
- mu [mean vlaue]

### Probability mass function
$$ f(x,\mu) = Pr(X = x) =  e^{-\mu} \frac{\mu^x}{x!}$$

### cumulative proability density
$$ F(x; \mu) = e^{-\mu}\sum_{i=0}^{\lfloor x \rfloor} \frac{\mu^i}{i!}$$ 

```python
# generate random numbers
data_poisson = poisson.rvs(mu=4, size=1000)
```

```python
ax = sns.histplot(data_poisson,
                  kde=True,
                  color='skyblue')
ax.set(xlabel='Poisson Distribution', ylabel='Frequency');
```

```python
# simulation
"""
assume selling pizza in 28 days
mean value of selling pizza is 12 per day
"""
mu = 12
n = 28
x_vals = np.arange(n)
mu, var = poisson.stats(mu)
poisson_dist = [poisson.pmf(x, mu) for x in x_vals]
print("x\t P(X=x)")
for key, values in enumerate(poisson_dist):
    print(f"{key} \t {values}")
print(f"mean is {mu} and variance is {var}")
```

## Probabitliy mass distribution 

```python
plt.bar(x_vals, poisson_dist);
```

## Probability density function

```python
poisson_cdf = [poisson.cdf(x, mu) for x in x_vals]
plt.bar(x_vals, poisson_cdf);
```

```python
# likelihood of exactly 5 pizza will sold out in 1st day
# assume mean value for 28 days is 12
p_5_pizza_will_sold_out = poisson.pmf(12, 5)
display(Math(r'P(X=5) = {:.4f}'.format(p_5_pizza_will_sold_out)))
```

```python
# likelihood of at least 12 pizza will sold out
# assume mean value for 28 days is 12
p_at_least_12_pizzia_will_sold_out = 1 - poisson.cdf(12, 11)
display(Math(r'P(X \geq 12) = 1 - P(X \leq 11)'))
display(Math(r'P(X \geq 12) = {:.4f}'.format(p_at_least_12_pizzia_will_sold_out)))
```

```python
# at most 12 pizza wil sold out
p_at_most_12_pizza_will_sold_out = poisson.cdf(12, 12)
display(Math(r'P(X \leq 12) = {:.4f}'.format(p_at_most_12_pizza_will_sold_out)))
```
